## Data Visualization

### Let's compare `ShareWomen` and `Unemployment_rate`.

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[['ShareWomen', 'Unemployment_rate']], figsize=(12,8))
plt.show()
```

```python
sns.pairplot(recent_grads[['ShareWomen', 'Unemployment_rate']].dropna(), size=4)
sns.plt.show()
```

### Let's compare the share of men and women in engineering major.

```python
grads_eng_share = recent_grads[recent_grads['Major_category'] == 'Engineering']
```

```python
grads_eng_share['ShareMen'] = 1 - grads_eng_share['ShareWomen']
```

```python
grads_eng_share = grads_eng_share.set_index('Major')
```

```python
grads_eng_share = grads_eng_share[['ShareMen', 'ShareWomen']]
```

```python
grads_eng_share.head()
```

```python
grads_eng_share.plot(kind='bar', figsize=(12,12))
plt.show()
```

from the plot above, we found most students in engineering major are male.

### Let's compare the number of men and women in engineering major.

```python
grads_eng_num = recent_grads[recent_grads['Major_category'] == 'Engineering']
```

```python
grads_eng_num.head()
```

```python
grads_eng_num = grads_eng_num.set_index('Major')
```

```python
grads_eng_num = grads_eng_num[['Men', 'Women']]
```

```python
grads_eng_num.head()
```

```python
grads_eng_num.plot(kind='bar', stacked=True, figsize=(12,12))
plt.show()
```

from the stacked box plot above, we found that most students learn mechanical and electrical engineering, and the ratio between men and women.
